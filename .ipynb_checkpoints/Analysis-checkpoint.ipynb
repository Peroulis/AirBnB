```python
# library to make requests to a web page, and print the responce text
import requests 
import re
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np
from time import sleep
from datetime import datetime, timedelta
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
import sys

import urllib.parse as urlparse
from urllib.parse import parse_qs

import folium

from IPython.display import display, clear_output

import json

import webbrowser as web# to open urls instantly
```

```python
# Open saved scv file
data = pd.read_csv('appartments.csv')
```

```python
# create a map of appartments
m = folium.Map(location=[38.228921, 20.567190],tiles='Stamen Terrain')

for i in range(0,len(data)):
   folium.Marker(
      location=[data.iloc[i]['latitude'], data.iloc[i]['longitude']],
      popup = data.iloc[i]['url'],
   ).add_to(m)
m
```

```python
data.groupby('Beds')['PricePerNight'].mean().plot.bar(x='Beds',y='PricePerNight',label='Price Per Night',color = 'b')
data.groupby('Beds')['PricePerNight'].count().plot.bar(x='Beds',y='PricePerNight',label='# Apartments',color = 'g')
plt.legend()
plt.xlabel("Number of Beds")
plt.ylabel("Mean Price per Night")
plt.title("Mean Price per Night vs Number of Beds")
data.groupby('Beds')['PricePerNight'].count()#.plot(x='Beds',y='PricePerNight',kind='bar')
```

```python
# Appartments have mainly 1/2/3/4 beds. Thus 5/6 are filtered
data_filtered = data[data.Beds <= 4]
```

```python
data_filtered.mean()
```

```python
data_filtered.groupby('Visitors')['PricePerNight'].mean().plot.bar(x='Beds',y='PricePerNight',label='Price Per Night',color = 'b')
data_filtered.groupby('Visitors')['PricePerNight'].count().plot.bar(x='Beds',y='PricePerNight',label='# Apartments',color = 'g')
plt.legend()
plt.xlabel("Number of Visitors")
plt.ylabel("Mean Price per Night")
plt.title("Mean Price per Night vs Max Number of Visitors")
data_filtered.groupby('Visitors')['PricePerNight'].count()#.plot(x='Beds',y='PricePerNight',kind='bar')
```

```python
# Appartments which can have more than 6 visitors are too low. Thus these appartments could be excluded, but why should we
data_filtered = data_filtered[data_filtered.Visitors <= 5]
```

```python
def count_string(lst, regex_pattern):
    count = 0
    for elem in lst:
        print(lst)
        if re.search(regex_pattern, elem):
            count += 1
    return count

# apply the function to each list in the column
count_x = data_filtered.Amenities.apply(lambda x: count_string(x, 'Θέα'))
count_x.unique()

import seaborn as sns
sns.countplot(x=count_x)
plt.show()
```
