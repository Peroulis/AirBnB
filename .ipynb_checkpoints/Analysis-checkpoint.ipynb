```python
def Add_Feature(new_name,name,df):
    df[new_name] = df.Amenities.apply(lambda x: x.count(name))
    mean_values = df.groupby(new_name)['PricePerNight'].mean()#.plot.bar(x=new_name,y='PricePerNight',label='Price Per Night',color = 'b')
    count_values = df.groupby(new_name)['PricePerNight'].count()#.plot.bar(x=new_name,y='PricePerNight',label='# Apartments',color = 'g')
    
    x = count_values.index
    width = 0.35

    fig, ax = plt.subplots()
    
    rects1 = ax.bar(x - width/2, mean_values, width, label='Mean Cost Per Night')
    rects2 = ax.bar(x + width/2, count_values, width, label='# of Appartments')

    # Add labels and title
    ax.set_xticks(x)
    ax.set_ylabel('Value')
    ax.set_title('Two Series Bar Chart')
    ax.legend()
    # Show the plot
    plt.show()
  
    print(df.groupby(new_name)['PricePerNight'].mean(),df.groupby(new_name)['PricePerNight'].count())
    return mean_values,df
```

```python
new_feature = 'PlatesWashineMachine'
am = 'Πλυντήριο πιάτων'
m,amenities = Add_Feature(new_feature,am,amenities)
m[0]-m[1]
#if 100-m[0]/m[1]*100 > 5:
#    amenities = amenities[amenities[new_feature] != 0]
```
